```python
import requests, json, time

def steamspy(appid):
    r = requests.get("https://steamspy.com/api.php",
                     params={"request": "appdetails", "appid": appid}, timeout=10)
    return r.json()

def steamstore(appid, cc="us"):
    r = requests.get("https://store.steampowered.com/api/appdetails",
                     params={"appids": appid, "cc": cc, "l": "en"}, timeout=10)
    return r.json()[str(appid)]   # {success, data}

appid = 413150  # Stardew Valley
ss = steamspy(appid)
st = steamstore(appid)

print("=== SteamSpy ham ===")
print(json.dumps(ss, indent=2)[:1500])
print("\n=== Store success / alanlar ===")
print(st.get("success"), list(st.get("data", {}).keys()))
```
